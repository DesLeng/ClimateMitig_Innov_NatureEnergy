# NACI 2026: Deep-Dive Analysis - Transport (EV) & Renewables (Hydrogen)
This notebook replicates the logic from `05_deepdive_transport_ev.do` and `06_deepdive_renewables_hydrogen.do`.

```python
import asearch as asr
import pandas as pd
import plotly.express as px

# Connect to TIP Data
conn = asr.get_connection()

# Filter for Transport (Y02T) and Renewables (Y02E10)
df_ev = pd.read_sql("SELECT * FROM project_mitigation_enriched WHERE tech_4_digit = 'Y02T'", con=conn)
df_hydrogen = pd.read_sql("SELECT * FROM project_mitigation_enriched WHERE technology LIKE 'Y02E10%' OR technology LIKE 'Y02E60%'", con=conn)

print(f"EV records: {len(df_ev)}")
print(f"Hydrogen/Renewables records: {len(df_hydrogen)}")
```

## 1. EV Trend Analysis
Tracking the growth of Electric Vehicle technologies over time.

```python
ev_trends = df_ev.groupby('earliest_filing_year').size().reset_index(name='count')
fig = px.line(ev_trends, x='earliest_filing_year', y='count', title='Evolution of EV Innovation')
fig.show()
```

## 2. Hydrogen & Green Energy Transition
Analyzing the relative growth of Hydrogen technologies compared to traditional Renewables.

```python
hydro_trends = df_hydrogen.groupby(['earliest_filing_year', 'technology']).size().reset_index(name='count')
fig = px.area(hydro_trends, x='earliest_filing_year', y='count', color='technology', title='Hydrogen vs. Renewables Growth')
fig.show()
```
